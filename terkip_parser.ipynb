```python
### Useful functions for the main function

#At the end, the algorith will give a list of tokenized terkips. This unifier function unifies the tokens.
def unifier(lst):
  terkipler_unified = []
  for terkip in lst:
    type_terkip = type(terkip)

    if type_terkip == list:
      word = ''
      for kelime in terkip:
        if word == '':
          word = word + kelime
        else:
          word = word + ' ' + kelime
      terkipler_unified.append(word)  
    else:
      word = terkip
      terkipler_unified.append(word)

  return terkipler_unified 

import re
#These two regex functions checks whether a word matches the pattern
def tire_follow(word):
  return (re.findall(r'‘?\w+‘?-y?[ıiuü]', word)) != []
def int_tire(word):
  return (re.findall(r"‘?\w+‘?-[^yıiuü-]+", word)) != []

#Main function:
def terkip_parser(text):
  text = text.lower()
  token_text = text.split()


  izaf = ["ı", "i", "u", "ü"] #These letters connect parts of a terkip. For instance: "‘adl ü kerem"


  terkipler = []
  limit = len(token_text)

  index = 0
  while index < limit:

    token = token_text[index]
    if index < len(token_text)-1:
      muc1 = token_text[index+1]
    else:
      muc1 = ""
    
    terkip = []
    if int_tire(token) and not tire_follow(token):
      terkipler.append(token)
    elif tire_follow(token) or muc1 in izaf:
      if len(token_text) - index > 6:
        mucs = [token_text[token_text.index(token)+i] for i in range (1,7)]
      else:
        mucs = [token_text[token_text.index(token)+i] for i in range(1,len(token_text)-index)]
      terkip.extend([token,muc1])

      for muc in mucs[1:]:
        if muc in izaf:
          terkip.append(muc)
        elif tire_follow(muc):
          if tire_follow(mucs[mucs.index(muc)-1]) or mucs[mucs.index(muc)-1] in izaf:
              terkip.append(muc)
          else:
              break
        else:
          if mucs.index(muc) == 0:
            if tire_follow(muc1) in izaf:
              terkip.append(muc)
            else:
              break
          else:
            muc_pre = mucs[mucs.index(muc)-1]
            if tire_follow(muc_pre) or muc_pre in izaf:
              terkip.append(muc)
            else:
              break
    terkipler.append(terkip)
    lenter = len(terkip)
    if lenter != 0:
      index += lenter
    else:
      index += 1

  terkipler_filtered = [terkip for terkip in terkipler if terkip != []]
  return unifier(terkipler_filtered)
```
